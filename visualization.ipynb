```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
adidas = pd.read_csv('./Copy of Adidas_sales_datasets.xlsx - Data Sales Adidas.csv')
adidas.head()
```

```python
adidas.describe()
```

```python
adidas.info()
```

```python
# create a line plot using seaborn to display the regional distribution of total sales
plt.figure(figsize=(18, 24))
sns.lineplot(x='Region', y='Total Sales', data=adidas)
plt.title('Regional Distribution of Total Sales')
plt.ylabel('Total Sales')   
plt.xlabel('Region')
plt.show()
```

```python
# Create a bar chart to visualize the quantity of each product sold
plt.figure(figsize=(20, 24))
sns.barplot(x='Product', y='Units Sold',native_scale=True,data=adidas)
# ax.plot(markersize=1, color="r")
plt.title('Quantity of Each Product Sold')
plt.ylabel('Unit Sold')
plt.xlabel('Product')
plt.ylim(0,160)
plt.show()
```

```python
#  Use a scatter plot to explore this relationship between the price per unit and the operating profit

plt.figure(figsize=(20, 24))
sns.scatterplot(x='Price per Unit', y='Operating Profit', data=adidas)
plt.title('Relationship between the price per unit and the operating profit')
plt.ylabel('Operating Profit')
plt.xlabel('Pice per Unit')
plt.show()
```

```python
# Create a box plot to compare the distribution of operating margins for different sales methods
plt.figure(figsize=(20, 24))
sns.boxplot(x='Sales Method', y='Operating Margin', data=adidas)
plt.title('Distribution of Operating Margins for Different Sales Methods')
plt.ylabel('Operating Margin')
plt.xlabel('Sales Method')
plt.show()
```
